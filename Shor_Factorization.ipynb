```python
import numpy as np
import matplotlib.pyplot as plt
import math
import warnings
warnings.filterwarnings('ignore') # U^2 function encounters at least one divide by zero
```

```python
def x_calc(a,N,z):
    for x in range(10**4):
        if a**x % N == z and x!=0:
            return x
    raise Exception("Didnt find x")
        
def m_calc(Q,x_z,r):
    return 1+math.floor( (Q-x_z-1)/r )

def U_calc(n, theta):
    return np.sin((n+1)*np.arccos(theta)) / np.sin(np.arccos(theta))

def U2_calc(m, r, y, Q):
    arg = np.cos(np.pi*r*y/Q)
    return ( U_calc(m, arg) )**2
```

```python
N = 15
Q = 256
```

```python
a = 7
z = 4
r = 4

x_z = x_calc(a,N,z)
m = m_calc(Q,x_z,r)

print('x_z =', x_z)
print('m =', m)

fig = plt.figure(figsize=(12,6))
plt.xlabel('y', fontsize=25)
plt.ylabel('$U_{m-1}^2 cos(\pi r y /Q)$', fontsize=25)
plt.tick_params(labelsize=15)

ys = np.arange(1,Q+1,1)
verts = U2_calc(m, r, ys, Q)
plt.plot(ys, verts)
#plt.savefig('plot1.png'); plt.show()

peaks = np.array([])
for i in range(len(verts)):
    if verts[i] > 10:
        peaks = np.append(peaks, i)
print('peak 1 =', peaks[0])
print('peak 2 =', peaks[1])
print('distance between peaks =', peaks[1]-peaks[0])
print('r = 2^(8+1) / (distance between peaks) = ', 2*Q/(peaks[1]-peaks[0]) )
```
