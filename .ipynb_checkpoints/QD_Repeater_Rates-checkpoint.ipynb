```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n = 3
L_a = 22
c = 2.0 * 10**5
p = 0.9
n_d = 0.9

L_vals = np.arange(200,1001,1)

#######################################################
# rare earth
p_m = p * n_d
RE_rate_vals = []

for L in L_vals:
    L_0 = L / (2**n)
    n_t = np.exp(-(L_0)/(2.0*L_a))
    p_t = 0.5 * n_t**2 * p**2 * n_d**2
    p_0 = ((1.0/p_t) + (1.0/(p_t * p_m**2)))**(-1)
    p_s = p**2 * n_d**2
    T = (3.0/2)**(n-1) * (L_0 / c) / ( p_0 * p_s**n )
    rate = 1.0 / T
    RE_rate_vals = np.append(RE_rate_vals, rate)
    
#######################################################
# DLCZ
DLCZ_rate_vals = []

for L in L_vals:
    L_0 = L / (2**n)
    n_t = np.exp(-(L_0)/(2.0*L_a))   
    p_t = 0.5 * n_t**2 * p**2 * n_d**2
    p_s = p**3 * n_d**3
    n_m = 0.9
    eta = n_d * n_m
    T = 3**(n+1) * (L_0/c) * (2-eta)*(4-(3*eta))*(8-(7*eta)) / (n_d * n_t * p * eta**(n+2))
    rate = 1.0 / T
    DLCZ_rate_vals = np.append(DLCZ_rate_vals, rate)

#######################################################
# Direct Transmision
direct_rate_vals = []

for L in L_vals:
    n_t = np.exp(-L/L_a)
    rate = 10 * 10**9 * n_t
    direct_rate_vals = np.append(direct_rate_vals, rate)
      
#######################################################
# QD
QD_rate_vals = []

for L in L_vals:
    L_0 = L / (2**n)
    n_t = np.exp(-(L_0)/(2.0*L_a))   
    p_t = 0.5 * n_t**2 * p**2 * n_d**2
    p_s = p**3 * n_d**3
    T = 0.032 + (3.0/2)**n * (L_0 / c) / ( p_t * p_s**n )
    rate = 1.0 / T
    QD_rate_vals = np.append(QD_rate_vals, rate)

#######################################################
# plot
fig = plt.figure(figsize=(10,6))
plt.xlabel('Distance [km]', fontsize=25)
plt.ylabel('Distribution rate [Hz]', fontsize=25)
plt.yscale('log')
plt.xlim(200,1000); plt.ylim(0.001, 5*10**4)
plt.xticks([200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000], \
      [r'$200$', r'', r'', r'', r'$400$', r'', r'', r'', r'$600$', r'', r'', r'', r'$800$', r'', r'', r'', r'$1000$'])
plt.yticks([0.01, 0.1, 1, 10, 100, 10**3, 10**4], [r'$0.01$', r'', r'$1$', r'', r'$100$', r'', r'$10^4$'])
plt.tick_params(labelsize=25)
plt.tick_params(direction='in', length=6, width=2, colors='k',
              grid_color='k', grid_alpha=0.5, which='major')
plt.tick_params(direction='in', length=0, width=0, colors='k',
              grid_color='k', grid_alpha=0.5, which='minor')

plt.plot(L_vals, QD_rate_vals, lw=4.5, c='purple', label='QDs')
plt.plot(L_vals, direct_rate_vals, lw=4.5, c='orange', label='Direct')
plt.plot(L_vals, RE_rate_vals, lw=4.5, c='g', label='Rare Earth')
plt.plot(L_vals, DLCZ_rate_vals, lw=4.5, c='b', label='DLCZ')

legend = plt.legend(loc=1, shadow=(True), title='Protocol', fontsize=22)
title = legend.get_title(); title.set_fontsize(20)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)
#plt.savefig('rates.png')
plt.show()
```
