```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# side
pos = np.array([0,1.1189,2.0486,2.5807,3.0631,3.4854]) # flat,99,98,97,96,95

delta_Es = [1.164,1.1564,1.152,1.150,1.146,1.141]
delta_Qs = [0.3058,0.6953,0.5962,0.5100,0.5494,0.5451]
ground_omegas = [103.40,45.83,53.64,63.15,59.03,59.53]
excited_omegas = [107.08,47.02,55.89,62.5,61.01,61.72]
Sfs = [1.1566,2.6505,2.2809,1.9644,2.1311,2.1162]
Xs = [1.9787e-7,8.6593e-14,9.6055e-13,1.1869e-10,4.4430e-11,4.9243e-11]
Ws = [0.0241,0.0104,0.0122,0.0116,0.0081,0.0097]
Gamma = [2197433,0.1,3,303,55,88]
```

```python
fig = plt.figure(figsize=(8,6))
plt.ylabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.xlabel('Compressive strain (%)', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)

lattice_vec = np.array([1,0.99,0.98,0.97,0.96,0.95])
plt.scatter((1-lattice_vec)*100, pos)

coef = np.polyfit((1-lattice_vec)*100, pos, 2)
#print(coef) # [-0.09414643  1.150495    0.03622143]
# height = 0.036 + (1.15x) + (-0.09414643)
poly1d_fn = np.poly1d(coef)  
x = np.linspace(0,5.3,1000)
plt.plot(x, poly1d_fn(x))

plt.grid(linestyle='dashed')
plt.show()
```

```python
# transition energies
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$\Delta E$ [eV]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.ylim(0,2)
plt.grid(linestyle='dashed')
plt.scatter(pos,delta_Es, s=70); plt.show()
```

```python
# delta Q's
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$\Delta Q$ [amu$^{1/2}$ $\AA$]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.ylim(0,1)
plt.grid(linestyle='dashed')
plt.scatter(pos,delta_Qs, s=70); plt.show()
```

```python
# effective frequencies
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$\Omega$ [meV]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.ylim(40,120)
plt.scatter(pos,ground_omegas, s=50, label='$\Omega_g$')
plt.scatter(pos,excited_omegas, s=50, label='$\Omega_e$')

legend = plt.legend(loc=1, shadow=(True), fontsize=22)
frame = legend.get_frame()
frame.set_facecolor('#f9f9f9')
frame.set_alpha(0.6)
plt.grid(linestyle='dashed')
plt.show()
```

```python
# HR factors
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('HR factor ($S_f$)', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.ylim(0.8,3)
plt.scatter(pos,Sfs, s=70)
plt.grid(linestyle='dashed')
plt.show()
```

```python
# Phonon coupling
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$X_{if}$ [amu $\AA^2$/eV]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.yscale("log")
plt.scatter(pos,Xs, s=70)

plt.grid(linestyle='dashed')
plt.show()
```

```python
# Electronic coupling
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$W_{if}$ [eV/amu$^{1/2}$ $\AA$]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.scatter(pos,Ws, s=70)
plt.grid(linestyle='dashed')
plt.show()
```

```python
# NR rates
fig = plt.figure(figsize=(8,5))
plt.xlabel('Wrinkle height [$\AA$]', fontsize=35, labelpad=10)
plt.ylabel('$\Gamma_{nr}$ [Hz]', fontsize=35, labelpad=20)
plt.tick_params(labelsize=25)
plt.yscale("log")
plt.scatter(pos,Gamma, s=70)
plt.grid(linestyle='dashed')
plt.show()
```

```python
from scipy.optimize import curve_fit
from nonrad.nonrad import AMU2KG, ANGS2M, EV2J, HBAR

bond_len = np.arange(1.00,1.60,0.05)
energy = np.array([-38.84540837,-38.98448801,-39.07905601,-39.13957510,-39.17413928,-39.18914095,-39.18955048,-39.17921375,\
         -39.15662844,-39.13773081,-39.11527702,-39.08163051,-39.07060455])*13.605698066

plt.scatter(bond_len, energy, c='k')
plt.scatter(bond_len[4:9], energy[4:9], c='k')


def f(x, omega, dE,d):
    return 0.5 * omega**2 * (x-d)**2 + dE

bounds = (-np.inf, np.inf)
popt_ground, _ = curve_fit(f, bond_len[4:8], energy[4:8], bounds=bounds)
q = np.linspace(1.0, 1.6, 1000)
plt.plot(q, f(q, *popt_ground), lw=3, c='r')
plt.scatter(bond_len[4:9], energy[4:9], c='r')

reduced_mass = 1.0/ ((1/10.811) + (1/14.0067))
omega = HBAR * popt_ground[0] * np.sqrt(EV2J / (ANGS2M**2 * AMU2KG * reduced_mass) )
print('Frequency =', omega*10**3 , 'meV')
```
